### Bucle While

image.png

````python
## Vamos fazer alguns exemplos
## Vamos fazer o exercício para número, mas dentro de uma lista de números
## Vamos parar o ciclo dos loopings, assim que encontrarmos o par

is_even = False
index = 0
numbers = [11, 5, 13, 8, 34]

while not is_even:
    if numbers[index] % 2 == 0:
        is_even = True
        print("Even number found: " + str(numbers[index]))
    
    index += 1
````

````python
## Lembre-se que a condição pode ser qualquer expressão
## Que avalia uma condição lógica, você pode usar os operadores relacionais ou os lógicos


## O que você acha que acontece neste caso?
## Ele será executado para sempre, então, após um momento, pare de executar manualmente
while "OPS":
    print("We are inside the loop")
````

````python
## Vamos usar o mesmo exemplo que usamos para o loop 'while',
## Só que desta vez vamos iterar a lista usando o 'for' ... 'in'

numbers = [79, 3, 4, 5, 6]
for number in numbers:
    if number % 2 == 0:
        print("The number is even: " + str(numbers[index]))
        break
        
## A palavra break é para quebrar a iteração do laço
````

````python
## Vejamos o seguinte exemplo:

for number in "numbers":
    print("We are in the loop")
    
## Você percebeu que ele entra no ciclo de iteração?
## Podemos iterar uma string, mas o que exatamente vemos:

for letter in "numbers":
    print(letter)
    
## Podemos ver cada letra dentro da string
````
